# Vanishing and exploding gradient

## Exploding gradient

```python
import numpy as np
```

```python
x = np.random.rand((256))
for i in range(100):
    w = np.random.rand(256, 256)
    x = x@w
    print(x.mean(), x.std())
```

## Vanishing gradient

```python
x = np.random.rand(256)

for i in range(100):
    w = np.random.rand(256, 256) * 0.001
    x = x@w
    print(x.mean(), x.std())
```

# Xavier Problem Solution

## Normal Distribution

```python
x = np.random.rand(256)

for i in range(100):
    w = np.random.rand(256, 256) * (1/np.sqrt(256))
    x = x@w
    print(x.mean(), x.std())
```

## Uniform distribution

```python
x = np.random.rand(256)

for i in range(100):
    w = np.random.uniform(-1, 1, (256, 256)) * (1/np.sqrt(256))
    x = x@w
    print(x.mean(), x.std())
```

Although problem is reduced, it still is present and here comes the solution

```python
x = np.random.rand(256)

for i in range(100):
    w = np.random.uniform(-1, 1, (256, 256)) * (np.sqrt(6)/np.sqrt(256+256))
    x = x@w
    print(x.mean(), x.std())
```
